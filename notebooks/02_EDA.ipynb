```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("data/cleaned_sales_data.csv")
```

```python
sns.scatterplot(x="price", y="units_sold", data=df)
plt.title("Price vs Demand")
plt.show()
```

```python
#Customers buy less when price rises (Demand is price-sensitive)
```

```python
df.groupby("month")["units_sold"].sum().plot(kind="bar")
plt.title("Monthly Demand")
plt.show()
```

```python
#Prices could be higher in peak months
```

```python
sns.boxplot(x="discount", y="units_sold", data=df)
plt.title("Discount vs Demand")
plt.show()
```

```python
df.groupby("is_weekend")["units_sold"].mean()
```

```python
#This shows customers buy roughly the same quantity on weekdays and weekends.
```
